4. Write a program that takes a number n as input from the user and generates the first 
n terms of the series formed by squaring the natural numbers. 
Sample output 
Enter a number: 6 
The first 6 terms of the series are: 
1 4 9 16 25 36 

```python

n = int(input("Enter your number: "))

print (f"The first {n} trems of serise:", end=" ")
i = 1
while i <= 1:
    print(i * i, end="")
    i += 1
print()
```

5. Write a program that prompts the user to input a number and prints its multiplication 
table.  

```python

num = int(input("Enter a multiplication table number: "))

print(f"\nMultiplication Table of {num}:")

for a in range(1, 11):
    print(f"{num} x {a} = {num * a}")
```

6. Write a program that prompts the user to enter a number n, and then prints all the 
odd numbers between 1 and n.   

```python
num = int(input("Enter your number"))

for a in range (1 , num + 1):
    if a % 2 != 0:
        print(a , end="")
print()
```

9. Write a program that prompts the user to enter a number and repeats this process 5

running total. 
- Sample Output: 
- Enter a number: 10 
- Enter a number: 15 
- Enter a number: 35 
- Enter a number: 40 
- Enter a number: 50 
- The final running total is: 150   

```python

running_total = 0

for a in range(5):
    num = int(input("ENter your number"))
    running_total += num
print(f"The final runnig_total number is: {running_total}")
```

10. Write a Python program that:

1- Creates a list named clean_countries containing the following countries:
[Bangladesh
India
Pakistan
China]

2- Repeatedly asks the user to enter a country name.

3- Checks whether the entered country exists in the clean_countries list.

4-If the country is found in the list:

Display the message:

<country_name> is in clean countries list

Stop the program.

5-If the country is not found, keep asking the user to enter another country.

```python

clean_countries = ["pakistan", "India", "china", "bangladash"]

while True:
    user_country = input("Enter Country Name")

    if user_country in clean_countries:
        print(f"{user_country} in list in clean_countries list")
        break
    else:
        print("Your country is not found. please try again")
```
